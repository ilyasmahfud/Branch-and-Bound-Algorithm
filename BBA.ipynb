```python

import math 
maxsize = float('inf') 

# fungsi untuk menyalin solusi sementara ke solusi akhir
def copyToFinal(curr_path): 
    final_path[:N + 1] = curr_path[:] 
    final_path[N] = curr_path[0] 

# fungsi untuk mencari edge minimum total sampai vertex terakhir
def firstMin(adj, i): 
    min = maxsize 
    for k in range(N): 
        if adj[i][k] < min and i != k: 
            min = adj[i][k] 

    return min

# fungsi untuk mencari edge minimum total kedua sampai vertex terakhir
def secondMin(adj, i): 
    first, second = maxsize, maxsize 
    for j in range(N): 
        if i == j: 
            continue
        if adj[i][j] <= first: 
            second = first 
            first = adj[i][j] 

        elif(adj[i][j] <= second and
            adj[i][j] != first): 
            second = adj[i][j] 

    return second 


# curr_bound -> batas bawah dari simpul akhir
# curr_weight-> menyimpan bobot yang sudah ditempuh
# level -> lokasi simpul sementara ketika bergerak/berpindah dalam
# pohon pencarian(search tree)
# curr_path[] -> menyimppan solusi sementara
# yang nantinya akan disimpan di penyelesaian_lintasan_akhir[] 
def TSPRec(adj, curr_bound, curr_weight, 
            level, curr_path, visited): 
    global final_res 
    
    # penyelesaian akhirnya adalah ketika kita telah mencapai level N
    # yang berartu kita telah melintasi semua vertex
    if level == N: 
        
        # periksa apakah ada edge dari vertex terakhir di jalur
        # jika tidak, kembali ke simpul pertama
        if adj[curr_path[level - 1]][curr_path[0]] != 0: 
            
            # curr_res memiliki bobot total dari solusi yang kita dapat
            curr_res = curr_weight + adj[curr_path[level - 1]][curr_path[0]] 
            if curr_res < final_res: 
                copyToFinal(curr_path) 
                final_res = curr_res 
        return

    # berikut ini adalah iterasi
    # untuk level itersi yang lain pada semua vertex 
    # untuk membuat search tree yang rekursif
    for i in range(N): 
        
        # pastikan vertex selanjutnya tidak sama atau belum dikunjungi 
        if (adj[curr_path[level-1]][i] != 0 and visited[i] == False): 
            temp = curr_bound 
            curr_weight += adj[curr_path[level - 1]][i] 

            # different computation of curr_bound 
            # for level 2 from the other levels 
            if level == 1: 
                curr_bound -= ((firstMin(adj, curr_path[level - 1]) +
                                firstMin(adj, i)) / 2) 
            else: 
                curr_bound -= ((secondMin(adj, curr_path[level - 1]) +
                                firstMin(adj, i)) / 2) 

            # curr_bound + curr_weight adalah batas bawah dari vertex
            # yang sudah dikunjungi
            # jika batas bawah yang sekarang < final_res, 
            # kita bisa lanjut ke vertex setelahnya
            if curr_bound + curr_weight < final_res: 
                curr_path[level] = i 
                visited[i] = True
            
                # panggil fungsi TSPRec untuk lanjut ke simpul selanjutnya
                TSPRec(adj, curr_bound, curr_weight, 
                    level + 1, curr_path, visited) 

            # jika tidak, maka kita reset semua perubahan di curr_weight
            # dan curr_bound dan ganti ke vertex yang lain
            curr_weight -= adj[curr_path[level - 1]][i] 
            curr_bound = temp 

            
            # untuk mereset array yang sudah dikunjungi
            visited = [False] * len(visited) 
            for j in range(level): 
                if curr_path[j] != -1: 
                    visited[curr_path[j]] = True

# Berikut ini adalah fungsi untuk mencari penyelesaian rute tercepat
def TSP(adj): 

    # Hitung batas awal pertama untuk vertex awal
    # gunakan rumus 1/2 * (sum of first min + second min) untuk semua edge. 
    # juga inisialisasi jalur_sementara dan daftar yang sudah dikunjungi
    curr_bound = 0
    curr_path = [-1] * (N + 1) 
    visited = [False] * N 

    # menghitung batas awal
    for i in range(N): 
        curr_bound += (firstMin(adj, i) +
                    secondMin(adj, i)) 

    # Rounding off the lower bound to an integer 
    curr_bound = math.ceil(curr_bound / 2) 

    # kita mulai di vertex awal, sehingga vertex pertama 
    # di curr_path[] = 0
    visited[0] = True
    curr_path[0] = 0

    # memanggil fungsi TSPRec untuk curr_weight 
    # yang sama dengan 0 di simpul pertama
    TSPRec(adj, curr_bound, 0, 1, curr_path, visited) 


# matrix adjancent dari graf yang ada
adj = [[0, 550, 1200, 1200, 1500, 2000, 2800, 2800, 3200, 4000, 5800], 
       [550, 0, 700, 750, 850, 1400, 2300, 2150, 2750, 3550, 5450], 
       [1200, 700, 0, 900, 1600, 700, 1600, 2300, 2900, 3700, 5800], 
       [1200, 750, 900, 0, 700, 1200, 2300, 1400, 2000, 2800, 4900],
       [1500, 850, 1600, 700, 0, 1900, 3000, 1200, 1700, 2500, 4700],
       [2000, 1900, 700, 1200, 1900, 0, 1600, 2300, 1500, 2300, 4300],
       [2800, 1600, 1600, 2800, 3200, 1600, 0, 2700, 1200, 2000, 2400],
       [2800, 2150, 2300, 1400, 1200, 2300, 2700, 0, 1500, 1800, 3700],
       [3200, 27500, 2900, 2000, 2000, 1500, 1200, 1500, 0, 800, 2400],
       [4000, 3550, 4100, 3200, 2800, 2300, 2000, 1800, 800, 0, 1900],
       [5800, 5450, 5800, 4900, 4700, 4300, 2400, 3700, 2400, 1900,0]] 
N = 11

# final_path[] menyimpan penyelesaian akhir dari rute terpendek
final_path = [None] * (N + 1) 

# visited[] menyimpan jalur dari vertex yang sudah dikunjungi
visited = [False] * N 

# menyimpan bobot minimum akhir dari rute terpendek
final_res = maxsize 

TSP(adj) 

print("0 = Witel Solo")
print("1 = ODP A")
print("2 = ODP B")
print("3 = ODP C")
print("4 = ODP D")
print("5 = ODP E")
print("6 = ODP F")
print("7 = ODP G")
print("8 = ODP H")
print("9 = ODP I")
print("10 = ODP J")

print("Bobot Minimum      :", final_res) 
print("Jalur yang diambil : ", end = ' ') 
for i in range(N + 1): 
    print(final_path[i], end = ' ') 

```
